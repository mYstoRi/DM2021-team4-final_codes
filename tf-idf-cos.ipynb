```python
doc = []
doc_index = []
for i in range(1402):
    try:
        with open('data\\Train\\dataTrainComplete\\' + str(i) + '.txt', 'r', encoding='UTF-8') as f:
            doc.append(''.join(list(f)))
            doc_index.append(i)
    except:
        pass
```

```python
tokens = []
for d in doc:
    temp = []
    res = jieba.cut(d)
    for word in res:
        temp.append(word)
    tokens.append(temp.copy())
tokens[9]
```

```python
clean_tokens = []
for d in tokens:
    temp = []
    for word in d:
        if word > u'\u4e00' and word < u'\u9fff':
            temp.append(word)
    clean_tokens.append(temp.copy())
clean_tokens[0:2]
```

```python
for docu in range(len(clean_tokens)):
    for word in range(len(clean_tokens[docu])):
        for feature in cropf+chemf+pestf:
            if clean_tokens[docu][word] in feature:
                clean_tokens[docu][word] = feature[0]
```

```python
tokens_with_space = []
for d in clean_tokens:
    doc_with_space='' #這部分是讓中文文本看起來比較像英文，方便vectorizer分析
    for token in d:
        doc_with_space += token
        doc_with_space += ' '
    tokens_with_space.append(doc_with_space)
tokens_with_space[0]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(tokens_with_space)
print(tfidf_matrix.toarray())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosinescore = cosine_similarity(tfidf_matrix,tfidf_matrix)
```

```python
one_directional = 0
two_directional = 0
total = len(answersheet)
for rel in answersheet:
    if [rel[1], rel[0]] in answersheet:
        two_directional += 1
    else:
        one_directional += 1
print(one_directional)
print(two_directional)
print(total)
print(one_directional/total)
print(two_directional/total)
```

```python
def f1score(gen, ans):
    tp = 0
    fp = 0
    fn = 0
    for pair in gen:
        if (pair in ans)|([pair[1],pair[0]] in ans):
            tp += 1
        else:
            fp += 1
    for pair in ans:
        if not ((pair in gen)|([pair[1],pair[0]] in gen)):
            fn += 1
    precision = tp/(tp+fp)
    recall = tp/(tp+fn)
    score = 2*(precision*recall/(precision+recall))
    return score
```

```python
for i in range(len(doc_index)):
    doc_index[i] = str(doc_index[i])
```

```python
criteria = 0.6
dx = 0.01
f1last = 0

for z in range(100):
    gen_pairs = []
    gen_pairsdx = []
    for i in range(len(cosinescore)):
        for j in range(i):
            if cosinescore[i][j] > criteria:
                gen_pairs.append([doc_index[i],doc_index[j]])
            if cosinescore[i][j] > (criteria+dx):
                gen_pairsdx.append([doc_index[i],doc_index[j]])
    f1 = f1score(gen_pairs, answersheet)
    f1dx = f1score(gen_pairsdx, answersheet)
    criteria += ((f1dx-f1)/dx)*0.2*((100-z)/100)
    print('iteration : ' + str(z+1))
    print('criteria : ' + str(criteria))
    print('f1 : ' + str(f1))
```
