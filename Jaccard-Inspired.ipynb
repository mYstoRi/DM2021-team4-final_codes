```python
def jaccard(testdoc, refdoc):
    tp, fp, fn = 0, 0, 0
    for keyword in testdoc:
        if keyword in refdoc:
            tp += 1
        else:
            fp += 1
    for keyword in refdoc:
        if keyword not in testdoc:
            fn += 1
    return (tp, fp, fn)
```

```python
def f1score(gen, ans):
    tp = 0
    fp = 0
    fn = 0
    for pair in gen:
        if (pair in ans):
            tp += 1
        else:
            fp += 1
    for pair in ans:
        if not (pair in gen):
            fn += 1
    precision = tp/(tp+fp)
    recall = tp/(tp+fn)
    score = 2*(precision*recall/(precision+recall))
    return precision, recall, score
```
